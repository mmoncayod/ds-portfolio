# Cutting sequences

## Phred quality system

+ Scores of 20 or more: High quality, with an error rate of 1% or less.
+ Scores of 30 or more: Very high quality, with an error rate of 0.1% or less.
+ Scores below 20: Low to moderate quality, with an error rate greater than 1%.

Phred quality scores tell you the reliability of each base read during sequencing. By trimming sequences based on these scores, you are removing sections of sequences that have a high probability of being in error, which can distort your results during alignment and subsequent analyses.




```python
from Bio import SeqIO

for ab1_file in ab1_files:
    file_path = os.path.join(ab1_directory, ab1_file)
    record = SeqIO.read(file_path, 'abi')
    print(f"ID: {record.id}")
    print(f"Sequence length: {len(record)}")
    # Quality punctuation print
    print(record.letter_annotations["phred_quality"])
```

# **Saving the trimmed sequences for alignment in Clutscal Omega**

```python
from Bio import SeqIO
import os
import re

# Function to extract information from file names
def extract_sample_info(filename):
    # Updated to match "UEA4dR" for reverse instead of "UEA3dR"
    match = re.match(r"(\d+)(UEA3[EV])-UEA(3F|4dR)\.ab1", filename)
    if match:
        sample_number = int(match.group(1))
        population = match.group(2)[-1]  # 'E' o 'V'
        direction = 'F' if match.group(3) == '3F' else 'R'  # 'F' for forward, 'R' for reverse
        return (sample_number, population, direction)
    else:
        return None

# Create a list of tuples containing the extracted information and the file name
file_info = []
for ab1_file in ab1_files:
    info = extract_sample_info(ab1_file)
    if info:
        file_info.append((info, ab1_file))

# Sort the list by sample number, population and address
file_info_sorted = sorted(file_info, key=lambda x: (x[0][0], x[0][1], x[0][2]))

quality_threshold = 20
min_length = 100
trimmed_sequences = {}

# Iterate over the sorted list and process each file
for info, filename in file_info_sorted:
    file_path = os.path.join(ab1_directory, filename)
    record = SeqIO.read(file_path, 'abi')
    start_trim = next((idx for idx, qual in enumerate(record.letter_annotations["phred_quality"]) if qual >= quality_threshold), None)
    end_trim = next((idx for idx, qual in enumerate(reversed(record.letter_annotations["phred_quality"])) if qual >= quality_threshold), None)

    if start_trim is not None and end_trim is not None:
        end_trim = len(record) - end_trim
        if end_trim - start_trim >= min_length:
            trimmed_seq = record.seq[start_trim:end_trim]
            trimmed_record = record[start_trim:end_trim]  # Create a new record with the clipped sequence
            # Add the clipped record to the dictionary with the key as sample and population number
            sample_key = (info[0], info[1])
            if sample_key not in trimmed_sequences:
                trimmed_sequences[sample_key] = {}
            trimmed_sequences[sample_key][info[2]] = trimmed_record
        else:
            print(f"Secuencia de {record.id} es demasiado corta despus del recorte.")
    else:
        print(f"Secuencia de {record.id} no se pudo recortar debido a baja calidad general.")

# Create FASTA files for each pair of forward and reverse sequences
for sample_key, seq_dict in trimmed_sequences.items():
    if 'F' in seq_dict and 'R' in seq_dict:
        fasta_output_path = f"/content/drive/My Drive/Data Science/D. Citri/data/processed_data/trimmed_seqs_for_alignment/{sample_key[0]}{sample_key[1]}_pair.fasta"
        with open(fasta_output_path, "w") as output_handle:
            SeqIO.write([seq_dict['F'], seq_dict['R']], output_handle, "fasta")
            print(f"Archivo FASTA guardado para la muestra {sample_key[0]} poblacin {sample_key[1]} en: {fasta_output_path}")

```

# Sequence aligning

+ First: Assemble forward and reverse reads to obtain complete sequences for each sample
+ Second: Align those sequences within each population to compare between samples - Biopython

```python
from Bio import AlignIO, SeqIO
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
import os

# Defines the directory where the alignment files are located and the output directory
alignment_directory = '/content/drive/My Drive/Data Science/D. Citri/data/processed_data/aligned_sequences/'
consensus_output_directory = '/content/drive/My Drive/Data Science/D. Citri/data/processed_data/consensus_sequences/total_consensus_sequences/'

# Make sure the output directory exists, if not, create it
if not os.path.exists(consensus_output_directory):
    os.makedirs(consensus_output_directory)

# List all alignment files in the directory
alignment_files = [f for f in os.listdir(alignment_directory) if f.endswith('.txt')]

# Function to create consensus sequences
def create_consensus_from_alignment(alignment):
    consensus_seq = []
    for i in range(alignment.get_alignment_length()):
        # Get the basics for this position
        bases = alignment[:, i]
        # If all bases in this position are equal, or if there is a gap ('-'), take that base
        if bases.count(bases[0]) == len(bases) or bases[0] == '-':
            consensus_seq.append(bases[0])
        else:
            # If not, enter an 'N' to indicate the discrepancy.
            consensus_seq.append('N')
    # Returns the consensus sequence as a string
    return ''.join(consensus_seq)

# Iterate over each alignment file and create the consensus sequence
for alignment_file in alignment_files:
    # Build the full path to the alignment file
    alignment_path = os.path.join(alignment_directory, alignment_file)

    # Load the alignment
    alignment = AlignIO.read(alignment_path, "fasta")

    # Crear la secuencia consenso
    consensus_sequence_str = create_consensus_from_alignment(alignment)

    # Create a SeqRecord object with the consensus sequence
    consensus_record = SeqRecord(Seq(consensus_sequence_str), id=alignment_file.replace('.txt', ''), description="Consenso para " + alignment_file)

    # Print the consensus sequence or save it to a file
    print(consensus_record.format("fasta"))

    # Save the consensus sequence to a new file in the output directory
    consensus_output_path = os.path.join(consensus_output_directory, alignment_file.replace('.txt', '_consensus.fasta'))
    SeqIO.write(consensus_record, consensus_output_path, "fasta")
    print(f"Archivo FASTA de consenso guardado en: {consensus_output_path}")

```

# **Getting a single fasta file for every 32 samples for population E and V**


```python
import os

# Directory containing your FASTA files for populations E and V
directory_E = '/content/drive/My Drive/Data Science/D. Citri/data/processed_data/consensus_sequences/E/'
directory_V = '/content/drive/My Drive/Data Science/D. Citri/data/processed_data/consensus_sequences/V/'

# Function to check and merge FASTA files
def verify_and_combine_files(directory, population_suffix):
    combined_fasta_path = os.path.join('/content/drive/My Drive/Data Science/D. Citri/data/processed_data/consensus_sequences/combined_fasta_for_E_V/', f'combined_{population_suffix}.fasta')
    print(f"Combinando archivos para la poblacin {population_suffix}...")

    with open(combined_fasta_path, 'w') as outfile:
        # Counter to know how many sequences were added
        count = 0

        # List files and concatenate their contents if they are not empty
        for filename in sorted(os.listdir(directory)):
            if filename.endswith(f"{population_suffix}_FR_alligned_consensus.fasta"):
                filepath = os.path.join(directory, filename)
                if os.path.getsize(filepath) > 0:  # Make sure the file is not empty
                    with open(filepath, 'r') as infile:
                        contents = infile.read().strip()
                        if contents:
                            outfile.write(contents + '\n')
                            count += 1
                        else:
                            print(f"Archive {filename} is empty.")
                else:
                    print(f"Archive {filename} is empty or not accesible.")

        print(f"{count} sequences were added to the archive {combined_fasta_path}.")

# Run the function for both populations
verify_and_combine_files(directory_E, 'E')
verify_and_combine_files(directory_V, 'V')

```
